```python
import numpy as np
import scipy.special as sp
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
# Set up the domain of the plot
xplot = np.linspace(-15, 5, num=4096, endpoint=True)
# Set up the Ai(x) Bi(x)
Aix = sp.airy(xplot)[0]
Bix = sp.airy(xplot)[2]

# Plot the data
plt.rc('text', usetex=True)
plt.rc('font', family='serif', size="14")

#plt.figure(dpi=300)
plt.plot(xplot, Aix, 'r-', xplot, Bix, 'b--', xplot, 0.*xplot, 'k-')
plt.legend(['${\\rm Ai}(x)$', '${\\rm Bi}(x)$'], loc='best')
plt.xlim([-15,5])
plt.ylim([-.5,1])
plt.xlabel('$x$')
# plt.ylabel('')
plt.title('Airy functions');
# plt.savefig('airyfcns.png', facecolor='#ffffff')
```

```python
x = np.linspace(-15, 5, 2**10)

# plt.figure(dpi=300)
plt.plot(x, sp.airy(x)[1], 'r-', x, sp.airy(x)[3], 'b--', x, 0*x, 'k-')
plt.xlim([min(x), max(x)])
plt.ylim([-1.2,1.2])
plt.title('Derivatives of the Airy functions')
plt.xlabel('$x$')
plt.legend([r"${\rm Ai}'(x)$", r"${\rm Bi}'(x)$"]); # plt.savefig('fig/airyprime.svg', facecolor='#ffffff');
```
